## LanguageHaveWorkedWith

```python
# lets see which of the popular language in this survey is
```

```python
data.LanguageHaveWorkedWith
```

we see that all the language that a person has worked with are bundled in a single string separeted by semi colon, so lets first split and adding each string into columns of their own.

```python
# so lets create a language_worked_with dataframe of different language worked with.
language_worked_with = data.LanguageHaveWorkedWith.str.split(';', expand=True) 
```

above code hint : https://stackoverflow.com/questions/48958282/how-do-i-split-a-string-into-several-columns-in-a-dataframe-with-pandas-python

```python
language_worked_with.head()
```

```python
lang_work_w_col = [f'language_worked_with_{i+1}' for i in language_worked_with]
language_worked_with.columns = lang_work_w_col
```

```python
language_worked_with.head()
```

lets count the values of each language by iterating over each columns of language_worked_with and storing it in dictionary

```python
language_count = {}
for i in language_worked_with.columns:
    for j in language_worked_with[i]:
        if j is None:
            continue
        elif j not in language_count:
            language_count[j] = 1
        else:
            language_count[j] += 1
        
```

```python
language_count
```

```python
# lets reverse sort the language count according to values

lang_count_sort = {key: value for key, value in sorted(language_count.items(), key=lambda item: item[1], reverse=True)}
```

```python
lang_count_sort
```

```python
count = [count for count in lang_count_sort.values()]
language = [lang for lang in lang_count_sort.keys()]
```

```python
plt.figure(figsize=(15,10))
sns.barplot(y=language,x=count)
plt.title('Most used Programming languages')
plt.show()
```

So Python is the third most used language 

```python
language_worked_with.apply(pd.Series.value_counts).sum(axis=1).sort_values(ascending = False)
```

```python
# concate 'languge_worked_with' with data
data = pd.concat([data, language_worked_with], axis=1)
```
